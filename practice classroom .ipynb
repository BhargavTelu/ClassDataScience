```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans,AgglomerativeClustering,DBSCAN
from sklearn.preprocessing import StandardScaler,MinMaxScaler, RobustScaler
```

```python
data = pd.read_csv("iris.csv")
```

```python
data.head()
```

```python
features = data.iloc[:,0:4]
features
```

```python
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
Kmodel = KMeans(n_clusters=3)
Kmodel.fit_predict(scaled_features)
```

```python
Hmodel = AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='complete')
Hmodel.fit_predict(scaled_features)
```

```python
nc=range(2,11)
l=[]
for i in nc:
    
    model=KMeans(n_clusters=i)
    model.fit(scaled_features)
    l.append(model.inertia_)
```

```python
plt.figure(figsize=(10,4))
plt.style.use("seaborn-darkgrid")
plt.scatter(nc,l,s=100,color="black")
plt.plot(nc,l,linestyle="-.")
plt.xlabel("cluster")
plt.ylabel("Inertia")
plt.xticks(nc)
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import warnings
```

```python
df = pd.read_csv("iris.csv")
```

```python
df.head()
```

```python
label = LabelEncoder()
df["species"]=label.fit_transform(df["species"])
```

```python
df
```

```python
x_train,x_text, y_train,y_test = train_test_split(df.iloc[:,:-1],df.iloc[:,-1], test_size=0.2)
```

```python
species = []
for i in range (1,40,2):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train,y_train)
    pred = model.predict(x_text)
    species.append(accuracy_score(y_test,pred))
```

```python
species
```
